```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('melb_data.csv')
```

```python
#seperate the target from predictions
y = data.Price
X = data.drop(['Price'], axis = 1)

#divide the data into training and testing data
X_train_f, X_valid_f, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train_f.head()
```

```python
y_train.head()
```

```python
#selecting the categorical columns which have low cardinality(Cardinality means number of unique values in a column)

categorial_columns = [column_name for column_name in X_train_f.columns if X_train_f[column_name].nunique() < 10 and X_train_f[column_name].dtype=='object']

categorial_columns
```

```python
#selecting the numerical values
numerical_columns = [column_name for column_name in X_train_f.columns if X_train_f[column_name].dtype in ['int','float']]

numerical_columns
```

```python
required_cols = categorial_columns + numerical_columns
X_train = X_train_f[required_cols].copy()
X_valid = X_valid_f[required_cols].copy()
```

Pipelines are a simple way to keep your data preprocessing and modeling code organized. Specifically, a pipeline bundles preprocessing and modeling steps so you can use the whole bundle as if it were a single step.\n\n

Some benefits of using Pipelines are:\n
- Cleaner code
- easier to productionize
- fewer bugs

Full Pipeline is constructed in 3 steps:

Step 1: Define preprocessing steps
Pipeline bundles preprocessing steps and modeling steps together with the help of "ColumnTransformer" class 



```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder


#preprocessing for numerical data
numerical_transformer = SimpleImputer(strategy="constant")

#preprocessing for categorical data
categorical_transformers = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy="most_frequent")),
    ('onehot', OneHotEncoder(handle_unknown="ignore"))
])

#bundline preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers = [
        ('numerical', numerical_transformer, numerical_columns),
        ('categorical', categorical_transformers, categorial_columns)
    ]
)

```

Step 2: Define model

For this example, RandomForest regressor is used as the model

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=150, random_state=0)
```

Step 3: Create and Evaluate Pipeline

Use Pipeline class to define the pipeline that bundles the preprocessing and modeling steps. Few important points to remember:
- with pipeline, we can preprocess and model the data that fits in single line of code whereas without pipeline, seperate code for imputing, one hot encoding, model training takes place.
- With the pipeline, we supply the unprocessed features in X_valid to the predict() command, and the pipeline automatically preprocesses the features before generating predictions. (However, without a pipeline, we have to remember to preprocess the validation data before making predictions.)

```python
from sklearn.metrics import mean_absolute_error

my_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])

my_pipeline.fit(X_train, y_train)

#preprocessing the validation
preds = my_pipeline.predict(X_valid)

#evaluate the model
mse = mean_absolute_error(y_test, preds)
mse
```

Conclusion:

Pipelines are valuable for cleaning up machine learning code and avoiding errors, and are especially useful for workflows with sophisticated data preprocessing.
